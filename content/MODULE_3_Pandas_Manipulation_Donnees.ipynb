# Module 3 — Structures de données, fichiers CSV et premières données réelles

Dans les deux premiers modules, tu as appris les bases de Python : variables, types, listes, dictionnaires, conditions, boucles et fonctions. Maintenant, on commence à travailler comme un apprenant data : on organise des données, on lit un fichier CSV et on prépare les premières analyses.

Un fichier CSV est un fichier texte qui représente un tableau. Chaque ligne correspond souvent à une observation : une vente, un client, une commande, un produit. Chaque colonne décrit une information : date, client, montant, statut, canal, pays.

Ce module est très important parce que la data commence rarement avec de beaux graphiques. Elle commence avec des données à charger, vérifier, comprendre et structurer.

> **Objectif du module** : comprendre les structures de données utiles, lire un CSV, inspecter les lignes, vérifier les colonnes et préparer un premier tableau exploitable.


## 1. Pourquoi structurer les données ?

Structurer les données signifie les ranger d’une manière claire pour pouvoir les lire, les filtrer, les calculer et les expliquer.

Exemple simple : si tu as seulement la phrase “Awa a payé 29000 FCFA”, c’est lisible pour un humain, mais pas pratique pour faire des calculs. Si tu ranges cette information dans un dictionnaire, Python peut comprendre chaque partie.

```python
vente = {
    "client": "Awa",
    "montant": 29000,
    "paye": True
}
```

Ici, Python peut récupérer le client, additionner le montant ou vérifier si la vente est payée.

> **À retenir** : une donnée bien structurée rend l’analyse plus simple, plus rapide et moins risquée.


````python
vente = {
    "client": "Awa",
    "produit": "Formation Python",
    "montant": 29000,
    "paye": True
}

print("Client :", vente["client"])
print("Montant :", vente["montant"])
print("Paiement validé :", vente["paye"])
````

## 2. Une liste de dictionnaires : représenter plusieurs lignes

Une seule vente peut être représentée par un dictionnaire. Mais dans une vraie activité, on a plusieurs ventes. Pour représenter plusieurs lignes, on utilise souvent une **liste de dictionnaires**.

Chaque dictionnaire représente une ligne. La liste représente le tableau complet.

C’est une étape fondamentale avant Pandas, parce qu’un DataFrame Pandas ressemble beaucoup à cette idée : plusieurs lignes, avec les mêmes colonnes.

> **Image simple** : une liste de dictionnaires, c’est comme un petit tableau écrit en Python.


````python
ventes = [
    {"date": "2026-06-01", "client": "Awa", "montant": 29000, "canal": "WhatsApp", "paye": True},
    {"date": "2026-06-01", "client": "Moussa", "montant": 15000, "canal": "Facebook", "paye": False},
    {"date": "2026-06-02", "client": "Sarah", "montant": 45000, "canal": "TikTok", "paye": True},
]

print("Nombre de lignes :", len(ventes))
print("Première vente :", ventes[0])
````

## 3. Calculer à partir d’une liste de dictionnaires

Quand les données sont structurées, on peut les parcourir avec une boucle et calculer des indicateurs.

Un indicateur est un chiffre qui aide à comprendre une activité. Par exemple : chiffre d’affaires, nombre de clients, panier moyen, ventes payées, ventes à relancer.

Ici, on veut calculer le chiffre d’affaires uniquement sur les ventes payées. Cela oblige Python à vérifier la colonne `paye` avant d’additionner le montant.

> **Règle métier** : une vente non payée ne doit pas être comptée dans le chiffre d’affaires validé.


````python
chiffre_affaires = 0
clients_a_relancer = []

for vente in ventes:
    if vente["paye"]:
        chiffre_affaires = chiffre_affaires + vente["montant"]
    else:
        clients_a_relancer.append(vente["client"])

print("Chiffre d'affaires validé :", chiffre_affaires, "FCFA")
print("Clients à relancer :", clients_a_relancer)
````

## 4. Qu’est-ce qu’un fichier CSV ?

CSV signifie **Comma-Separated Values**, c’est-à-dire “valeurs séparées par des virgules”. En pratique, c’est un format très courant pour stocker des tableaux.

Un CSV peut venir d’un outil de vente, d’un formulaire, d’un export WhatsApp, d’une boutique en ligne ou d’un tableau Excel.

Exemple de contenu CSV :

```text
date,client,montant,canal,paye
2026-06-01,Awa,29000,WhatsApp,true
2026-06-01,Moussa,15000,Facebook,false
```

La première ligne contient les noms des colonnes. Les lignes suivantes contiennent les données.

> **À retenir** : un CSV est simple, mais il faut toujours vérifier les colonnes, les types et les valeurs manquantes.


````python
csv_texte = '''date,client,montant,canal,paye
2026-06-01,Awa,29000,WhatsApp,true
2026-06-01,Moussa,15000,Facebook,false
2026-06-02,Sarah,45000,TikTok,true'''

print(csv_texte)
````

## 5. Lire un CSV avec le module `csv`

Python possède un module intégré appelé `csv`. Un module est une boîte à outils déjà prête. Le module `csv` aide à lire les fichiers CSV ligne par ligne.

Dans cet exemple, on utilise `StringIO` pour simuler un fichier CSV directement dans le notebook. Dans un vrai projet, tu liras plutôt un fichier `.csv` enregistré dans ton dossier.

Le lecteur `csv.DictReader` transforme chaque ligne en dictionnaire. C’est très pratique, parce qu’on peut récupérer les valeurs avec le nom des colonnes.

> **Point important** : les valeurs lues depuis un CSV arrivent souvent comme du texte. Il faut convertir les montants avant de calculer.


````python
import csv
from io import StringIO

fichier = StringIO(csv_texte)
lecteur = csv.DictReader(fichier)

lignes = list(lecteur)

print("Nombre de lignes lues :", len(lignes))
print("Première ligne :", lignes[0])
print("Type du montant :", type(lignes[0]["montant"]))
````

## 6. Nettoyer les types après lecture

Après lecture du CSV, le montant est un texte (`str`). Pour calculer, on doit le convertir en nombre entier (`int`).

La colonne `paye` arrive aussi comme texte : `"true"` ou `"false"`. Pour faire des conditions propres, on la convertit en booléen : `True` ou `False`.

Cette étape s’appelle souvent **préparation des données**. Elle consiste à transformer les données brutes en données utilisables.

> **À retenir** : lire un fichier ne suffit pas. Il faut vérifier si les valeurs ont le bon type.


````python
ventes_preparees = []

for ligne in lignes:
    vente = {
        "date": ligne["date"],
        "client": ligne["client"],
        "montant": int(ligne["montant"]),
        "canal": ligne["canal"],
        "paye": ligne["paye"].lower() == "true",
    }
    ventes_preparees.append(vente)

print(ventes_preparees[0])
print("Type du montant préparé :", type(ventes_preparees[0]["montant"]))
````

## 7. Contrôler la qualité des données

Avant d’analyser, il faut faire des contrôles simples. Un contrôle qualité permet de détecter une erreur avant qu’elle fausse les conclusions.

Exemples de questions utiles :

- Est-ce que toutes les lignes ont un client ?
- Est-ce que tous les montants sont positifs ?
- Est-ce que la colonne `paye` est bien vraie ou fausse ?
- Est-ce que les canaux sont bien écrits ?

Dans un projet réel, ces contrôles évitent de présenter des chiffres faux au client ou à l’équipe.

> **Règle professionnelle** : un chiffre doit être vérifié avant d’être présenté.


````python
erreurs = []

for index, vente in enumerate(ventes_preparees, start=1):
    if vente["client"] == "":
        erreurs.append(f"Ligne {index} : client manquant")
    if vente["montant"] <= 0:
        erreurs.append(f"Ligne {index} : montant incorrect")

if erreurs:
    print("Erreurs trouvées :", erreurs)
else:
    print("Aucune erreur simple détectée")
````

## 8. Mini-projet guidé : premier résumé d’un CSV

Tu vas maintenant produire un résumé simple à partir de données structurées.

Le but n’est pas encore de faire une analyse complexe. Le but est de montrer que tu sais charger, préparer, vérifier et calculer.

Indicateurs à produire :

- chiffre d’affaires validé ;
- nombre total de ventes ;
- nombre de ventes payées ;
- panier moyen des ventes payées ;
- clients à relancer.

Ces indicateurs sont déjà utiles pour un petit business qui vend une formation, un coaching ou un service.


````python
total_valide = 0
ventes_payees = 0
clients_a_relancer = []

for vente in ventes_preparees:
    if vente["paye"]:
        total_valide += vente["montant"]
        ventes_payees += 1
    else:
        clients_a_relancer.append(vente["client"])

panier_moyen = total_valide / ventes_payees

print("Résumé du CSV")
print("--------------")
print("Ventes totales :", len(ventes_preparees))
print("Ventes payées :", ventes_payees)
print("Chiffre d'affaires validé :", total_valide, "FCFA")
print("Panier moyen payé :", panier_moyen, "FCFA")
print("Clients à relancer :", clients_a_relancer)
````

## Erreurs fréquentes à éviter

### 1. Croire que lire un CSV suffit

Lire un fichier n’est que le début. Il faut ensuite comprendre les colonnes, vérifier les types et contrôler les valeurs.

### 2. Calculer avec du texte

Si `montant` vaut `"29000"`, c’est du texte. Pour calculer, il faut faire `int(montant)`.

### 3. Oublier les lignes non payées

Dans un contexte business, toutes les commandes ne doivent pas forcément entrer dans le chiffre d’affaires validé. Il faut appliquer la règle métier.

### 4. Ne pas inspecter les premières lignes

Avant d’analyser, affiche toujours quelques lignes. Cela permet de voir si les données ont été chargées correctement.

### 5. Ne pas expliquer les résultats

Un résumé data doit être compréhensible. Ne donne pas seulement un chiffre : dis ce qu’il signifie.
