# Module 3 — Structures de données, fichiers CSV et premières données réelles

Ce module fait le pont entre Python pur et l'analyse de données réelle.

> **Objectif** : organiser des données avec listes et dictionnaires, comprendre un fichier CSV, lire un petit dataset et produire des premiers indicateurs.


## 1. Pourquoi structurer les données ?

Une donnée isolée est rarement suffisante. En data, on travaille avec des collections : plusieurs ventes, plusieurs clients, plusieurs produits.

| Structure | Usage |
|---|---|
| Liste | Plusieurs valeurs du même type. |
| Dictionnaire | Un objet complet avec plusieurs informations. |
| Liste de dictionnaires | Plusieurs objets complets. |
| CSV | Tableau sauvegardé dans un fichier. |


````python
clients = ["Awa", "Moussa", "Fatou"]

for client in clients:
    print("Client :", client)
````

## 2. Liste de dictionnaires : mini base de données

Une liste de dictionnaires permet de représenter plusieurs commandes.


````python
commandes = [
    {"client": "Awa", "montant": 29000, "statut": "payé"},
    {"client": "Moussa", "montant": 45000, "statut": "en attente"},
    {"client": "Fatou", "montant": 12000, "statut": "payé"},
]

for commande in commandes:
    print(commande["client"], "-", commande["montant"], "FCFA", "-", commande["statut"])
````

## 3. Filtrer avec une condition

Filtrer signifie garder seulement les éléments qui respectent une règle.

Exemple : garder les commandes payées.


````python
commandes_payees = []

for commande in commandes:
    if commande["statut"] == "payé":
        commandes_payees.append(commande)

print("Commandes payées :", len(commandes_payees))
for commande in commandes_payees:
    print(commande["client"], commande["montant"])
````

## 4. C'est quoi un fichier CSV ?

CSV signifie “Comma-Separated Values”. C'est un fichier texte qui représente un tableau.

Exemple :

```csv
client,produit,montant,statut
Awa,Formation Python,29000,payé
Moussa,Coaching Data,45000,en attente
```

> **À retenir** : un CSV est souvent le premier format de données réel qu'un analyste manipule.


## 5. Lire un CSV avec Python standard

Avant Pandas, il est utile de comprendre qu'un CSV peut être lu avec Python lui-même.


````python
import csv

texte_csv = """client,produit,montant,statut
Awa,Formation Python,29000,payé
Moussa,Coaching Data,45000,en attente
Fatou,Template Excel,12000,payé
"""

with open("mini_ventes.csv", "w", encoding="utf-8") as fichier:
    fichier.write(texte_csv)

with open("mini_ventes.csv", encoding="utf-8") as fichier:
    lecteur = csv.DictReader(fichier)
    for ligne in lecteur:
        print(ligne["client"], ligne["montant"], ligne["statut"])
````

## 6. Premiers indicateurs depuis le CSV

On transforme maintenant les lignes en chiffres utiles.


````python
import csv

ca = 0
commandes_payees = 0

with open("mini_ventes.csv", encoding="utf-8") as fichier:
    lecteur = csv.DictReader(fichier)
    for ligne in lecteur:
        if ligne["statut"] == "payé":
            ca = ca + int(ligne["montant"])
            commandes_payees = commandes_payees + 1

print("CA payé :", ca, "FCFA")
print("Commandes payées :", commandes_payees)
````

## 7. Introduction à Pandas

Pandas rend le travail sur tableaux beaucoup plus simple.

| Sans Pandas | Avec Pandas |
|---|---|
| Boucles manuelles. | Filtres directs. |
| Conversions manuelles. | Colonnes déjà structurées. |
| Beaucoup de code. | Code plus court et lisible. |


````python
import pandas as pd

df = pd.read_csv("mini_ventes.csv")
print(df)
````

## 8. Filtrer avec Pandas

Pandas permet de filtrer une table avec une condition claire.


````python
ventes_payees = df[df["statut"] == "payé"]
ca_paye = ventes_payees["montant"].sum()

print("CA payé :", ca_paye, "FCFA")
print("Nombre de ventes payées :", len(ventes_payees))
````

## 9. Erreurs fréquentes

| Erreur | Cause | Correction |
|---|---|---|
| Fichier introuvable | Mauvais dossier. | Vérifier `ls` / `dir`. |
| Mauvais nom de colonne | Colonne mal orthographiée. | Afficher `df.columns`. |
| Montant traité comme texte | Type non numérique. | Convertir avec `astype(int)`. |
| Encodage incorrect | Accents mal lus. | Utiliser `encoding="utf-8"`. |
