## 1. Variable : une information enregistrée avec un nom

Une variable est une zone de mémoire à laquelle on donne un nom. Elle permet de garder une information pour la réutiliser plus tard.

Imagine un cahier de caisse. Tu peux écrire : prix = 29000. Le mot `prix` devient l’étiquette, et `29000` devient la valeur gardée sous cette étiquette.

En Python, on crée une variable avec le signe `=`. Ce signe ne veut pas dire “égal” comme en mathématiques. Il veut dire : **range cette valeur dans ce nom**.

| Code | Sens |
|---|---|
| `prix = 29000` | on garde le prix dans la variable `prix` |
| `client = "Awa"` | on garde le nom du client dans `client` |
| `paye = True` | on garde l’information que le paiement est validé |

> **À retenir** : une variable doit avoir un nom clair. `chiffre_affaires` est beaucoup mieux que `x`, parce que le code doit être compréhensible par toi et par une autre personne.


````python
client = "Awa"
produit = "Formation Python"
prix = 29000
paye = True

print("Client :", client)
print("Produit :", produit)
print("Prix :", prix, "FCFA")
print("Paiement validé :", paye)
````

## 2. Les types de données : la nature de chaque information

Un type de donnée explique à Python quelle est la nature d’une valeur. C’est très important, parce que Python ne traite pas un texte, un nombre entier, un nombre décimal et une réponse oui/non de la même façon.

### `str` — chaîne de caractères

`str` vient de “string”. En français, on dit **chaîne de caractères**. C’est du texte : un nom, une ville, un produit, un email, un numéro de téléphone, une catégorie.

Une chaîne de caractères s’écrit entre guillemets : `"Awa"`, `"Formation Python"`, `"Abidjan"`.

Même si une valeur ressemble à un nombre, si elle est entre guillemets, Python la considère comme du texte. Par exemple `"29000"` est un texte, alors que `29000` est un nombre.

### `int` — nombre entier

`int` signifie integer, donc **nombre entier**. Il sert pour les quantités, les âges, les nombres de ventes, les stocks, les durées en jours.

### `float` — nombre décimal

`float` est un nombre avec une virgule, écrite avec un point en Python. Il sert pour les moyennes, les taux, les pourcentages, les notes, les prix décimaux.

### `bool` — vrai ou faux

`bool` signifie booléen. Il ne peut prendre que deux valeurs : `True` ou `False`. C’est utile pour représenter une décision : payé ou non payé, actif ou inactif, validé ou non validé.

| Type | Exemple | Signification data |
|---|---|---|
| `str` | `"Awa"` | texte : client, ville, produit |
| `int` | `29000` | nombre entier : prix, quantité, ventes |
| `float` | `4.5` | nombre décimal : moyenne, taux, note |
| `bool` | `True` | vrai/faux : payé, validé, actif |

> **Erreur fréquente** : confondre `"29000"` et `29000`. Le premier est un texte. Le second est un nombre calculable.


````python
nom_client = "Awa"
montant = 29000
note = 4.5
paiement_valide = True

print(type(nom_client))
print(type(montant))
print(type(note))
print(type(paiement_valide))
````

## 3. Calculs avec les nombres

Quand une valeur est un nombre, Python peut faire des calculs : addition, soustraction, multiplication, division, moyenne.

Dans un contexte business, c’est la base pour calculer un chiffre d’affaires, un panier moyen, un total de commande ou une remise.

| Opérateur | Rôle | Exemple |
|---|---|---|
| `+` | addition | `prix + livraison` |
| `-` | soustraction | `prix - remise` |
| `*` | multiplication | `prix * quantite` |
| `/` | division | `total / nombre_ventes` |

> **À retenir** : pour calculer, les données doivent être numériques (`int` ou `float`).


````python
prix = 29000
quantite = 3
livraison = 1000

total = prix * quantite + livraison

print("Total commande :", total, "FCFA")
````

## 4. Convertir une donnée : passer du texte au nombre

Dans les vrais fichiers de données, beaucoup de nombres arrivent sous forme de texte. Par exemple un prix peut venir comme `"29000"` au lieu de `29000`.

Avant de calculer, il faut convertir. Python propose des fonctions simples :

| Fonction | Rôle |
|---|---|
| `int()` | convertir en nombre entier |
| `float()` | convertir en nombre décimal |
| `str()` | convertir en texte |

La conversion est essentielle en data, parce qu’un mauvais type peut produire une erreur ou un résultat faux.

> **Exemple concret** : un prix écrit dans un formulaire web arrive souvent comme texte. Avant de multiplier ce prix par une quantité, on doit le convertir en nombre.


````python
prix_texte = "29000"
quantite = 2

prix_nombre = int(prix_texte)
total = prix_nombre * quantite

print("Total :", total, "FCFA")
````

## 5. Liste : plusieurs valeurs dans une seule variable

Une liste est une collection ordonnée de valeurs. Elle permet de stocker plusieurs éléments dans une même variable.

Dans la data, une liste peut représenter les ventes d’une journée, les noms des clients, les notes d’un test, les produits commandés ou les montants payés.

Une liste s’écrit avec des crochets `[]`. Les éléments sont séparés par des virgules.

Exemple : `ventes = [12000, 15000, 29000, 8000]`.

> **À retenir** : une liste sert quand tu as plusieurs valeurs du même genre à analyser ensemble.


````python
ventes = [12000, 15000, 29000, 8000]

chiffre_affaires = sum(ventes)
nombre_ventes = len(ventes)
panier_moyen = chiffre_affaires / nombre_ventes

print("Chiffre d'affaires :", chiffre_affaires, "FCFA")
print("Nombre de ventes :", nombre_ventes)
print("Panier moyen :", panier_moyen, "FCFA")
````

## 6. Dictionnaire : décrire une information complète

Un dictionnaire sert à garder des informations sous forme **clé → valeur**.

Une clé est le nom de l’information. La valeur est le contenu associé à cette clé.

Exemple :

```python
commande = {
    "client": "Awa",
    "produit": "Formation Python",
    "montant": 29000,
    "statut": "payé"
}
```

Dans la data, un dictionnaire est très utile pour représenter une ligne d’information : une commande, un client, un produit, une inscription, une transaction.


````python
commande = {
    "client": "Awa",
    "produit": "Formation Python",
    "montant": 29000,
    "statut": "payé"
}

print("Client :", commande["client"])
print("Montant :", commande["montant"], "FCFA")
print("Statut :", commande["statut"])
````

## 7. Condition : prendre une décision avec Python

Une condition permet à Python de choisir une action selon une situation.

En français, on dirait : si le paiement est validé, donner accès à la formation. Sinon, afficher une relance.

En Python, on utilise `if`, `else` et parfois `elif`.

- `if` signifie “si” ;
- `else` signifie “sinon” ;
- `elif` signifie “sinon si”.

Les conditions sont indispensables pour automatiser des règles métier : accès, validation, relance, classification, alerte.

> **Attention** : en Python, l’indentation est obligatoire. Les lignes qui appartiennent au `if` doivent être décalées avec des espaces.


````python
paiement_valide = True

if paiement_valide:
    print("Accès formation autorisé")
else:
    print("Paiement non validé : envoyer une relance")
````

## 8. Boucle : répéter une action automatiquement

Une boucle permet de répéter une action sur plusieurs éléments.

Quand tu as une liste de ventes, tu ne vas pas écrire un calcul ligne par ligne à la main. Tu demandes à Python de parcourir la liste et de faire l’action pour chaque élément.

La boucle la plus utilisée au début est `for`.

```python
for vente in ventes:
    print(vente)
```

> **À retenir** : la boucle économise du temps et réduit les erreurs manuelles.


````python
ventes = [12000, 15000, 29000, 8000]

for vente in ventes:
    print("Vente enregistrée :", vente, "FCFA")
````

## 9. Fonction : créer une action réutilisable

Une fonction est un bloc de code qui porte un nom et qu’on peut réutiliser plusieurs fois.

Si tu dois calculer le panier moyen plusieurs fois, tu ne dois pas recopier tout le code. Tu peux créer une fonction.

Une fonction commence avec `def`, puis le nom de la fonction, puis des paramètres entre parenthèses.

Dans la data, les fonctions rendent le travail plus propre, plus lisible et plus professionnel.


````python
def calculer_panier_moyen(ventes):
    total = sum(ventes)
    nombre = len(ventes)
    return total / nombre

ventes_lundi = [12000, 15000, 29000, 8000]
ventes_mardi = [10000, 18000, 22000]

print("Panier moyen lundi :", calculer_panier_moyen(ventes_lundi))
print("Panier moyen mardi :", calculer_panier_moyen(ventes_mardi))
````

## 10. Mini-projet guidé : analyser des ventes simples

Tu vas maintenant réunir les notions du module : variables, types, listes, dictionnaires, conditions, boucles et fonctions.

Objectif : analyser quelques commandes d’une petite activité de formation.

Tu dois être capable d’expliquer le chiffre d’affaires total, le nombre de commandes payées, le panier moyen et les commandes non payées à relancer.

Ce mini-projet est volontairement simple. Le but est de construire une base solide avant Pandas et les CSV.


````python
commandes = [
    {"client": "Awa", "montant": 29000, "paye": True},
    {"client": "Moussa", "montant": 29000, "paye": False},
    {"client": "Sarah", "montant": 45000, "paye": True},
    {"client": "Koffi", "montant": 15000, "paye": True},
]

total_paye = 0
nombre_payes = 0
clients_a_relancer = []

for commande in commandes:
    if commande["paye"]:
        total_paye = total_paye + commande["montant"]
        nombre_payes = nombre_payes + 1
    else:
        clients_a_relancer.append(commande["client"])

panier_moyen = total_paye / nombre_payes

print("Chiffre d'affaires validé :", total_paye, "FCFA")
print("Commandes payées :", nombre_payes)
print("Panier moyen :", panier_moyen, "FCFA")
print("Clients à relancer :", clients_a_relancer)
````

## Erreurs fréquentes à éviter

### 1. Oublier les guillemets autour du texte

`client = Awa` provoque une erreur, parce que Python croit que `Awa` est une variable. Il faut écrire `client = "Awa"`.

### 2. Confondre texte et nombre

`"29000" + "1000"` ne donne pas un vrai calcul. Ce sont deux textes. Pour calculer, il faut convertir en nombres.

### 3. Oublier l’indentation

Après `if`, `else`, `for` et `def`, les lignes du bloc doivent être décalées. Sans indentation correcte, Python ne comprend pas la structure.

### 4. Utiliser des noms trop vagues

`x`, `a`, `b` ne permettent pas de comprendre le code. Préfère `prix`, `ventes`, `chiffre_affaires`, `clients_a_relancer`.

### 5. Copier sans expliquer

Un bon apprenant data ne fait pas que copier le code. Il doit pouvoir expliquer chaque ligne avec ses mots.
