## 1. Variables : les boîtes de travail

Une variable garde une information avec un nom.

| Exemple | Sens |
|---|---|
| `prix = 29000` | Le prix vaut 29 000. |
| `client = "Awa"` | Le client s'appelle Awa. |
| `paye = True` | Le paiement est validé. |

> **À retenir** : un nom de variable doit être clair. `chiffre_affaires` est meilleur que `x`.


```python
client = "Awa"
produit = "Formation Python"
prix = 29000
paye = True

print("Client :", client)
print("Produit :", produit)
print("Prix :", prix, "FCFA")
print("Paiement validé :", paye)
```

## 2. Types de données essentiels

| Type | Exemple | Usage data |
|---|---|---|
| `str` | `"Awa"` | Nom, produit, ville. |
| `int` | `29000` | Quantité, âge, nombre de ventes. |
| `float` | `4.5` | Moyenne, taux, prix décimal. |
| `bool` | `True` | Oui/non, payé/non payé. |

> **Point d'attention** : `"29000"` est un texte, `29000` est un nombre. Python ne les traite pas pareil.


```python
prix = 29000
quantite = 3
total = prix * quantite

print("Total :", total, "FCFA")
print("Type du total :", type(total))
```

## 3. Listes : plusieurs valeurs dans une seule variable

Une liste permet de stocker plusieurs éléments.

Exemple : toutes les ventes d'une journée.

> **À toi de jouer** : ajoute une vente dans la liste, relance et observe le nouveau total.


```python
ventes = [12000, 15000, 29000, 8000]

chiffre_affaires = sum(ventes)
nombre_ventes = len(ventes)
panier_moyen = chiffre_affaires / nombre_ventes

print("Chiffre d'affaires :", chiffre_affaires, "FCFA")
print("Nombre de ventes :", nombre_ventes)
print("Panier moyen :", panier_moyen, "FCFA")
```

## 4. Dictionnaires : décrire un objet complet

Un dictionnaire garde des informations sous forme clé → valeur.

C'est très utile pour représenter un client, une commande ou un produit.


```python
commande = {
    "client": "Awa",
    "produit": "Formation Python",
    "montant": 29000,
    "statut": "payé"
}

print("Client :", commande["client"])
print("Montant :", commande["montant"], "FCFA")
print("Statut :", commande["statut"])
```

## 5. Conditions : décider avec Python

Une condition permet de prendre une décision.

Exemple business : si le paiement est validé, on donne accès à la formation ; sinon, on relance le client.


```python
paiement_valide = True

if paiement_valide:
    print("Accès formation accordé")
else:
    print("Relancer le client")
```

## 6. Boucles : répéter une action

Une boucle évite de répéter le même code à la main.

Exemple : afficher toutes les ventes et leur numéro.


```python
ventes = [12000, 15000, 29000]

for numero, montant in enumerate(ventes, start=1):
    print("Vente", numero, ":", montant, "FCFA")
```

## 7. Fonctions : rendre le code réutilisable

Une fonction regroupe une logique pour pouvoir la réutiliser.

> **À retenir** : si tu répètes le même calcul plusieurs fois, une fonction devient utile.


```python
def calculer_panier_moyen(ventes):
    total = sum(ventes)
    nombre = len(ventes)
    return total / nombre

ventes_jour = [12000, 15000, 29000, 8000]
resultat = calculer_panier_moyen(ventes_jour)

print("Panier moyen :", resultat, "FCFA")
```

## 8. Erreurs fréquentes

| Erreur | Cause probable | Correction |
|---|---|---|
| `NameError` | Variable mal écrite ou inexistante. | Vérifier le nom exact. |
| `TypeError` | Mélange texte/nombre. | Convertir avec `int()` ou `str()`. |
| `IndentationError` | Mauvais alignement. | Respecter les espaces sous `if`, `for`, `def`. |
| `ZeroDivisionError` | Division par zéro. | Vérifier que la liste n'est pas vide. |

> **Point d'attention** : en Python, l'indentation fait partie du langage.


## 9. Mini-défi métier

Crée une liste de ventes, calcule :

1. le chiffre d'affaires ;
2. le nombre de ventes ;
3. le panier moyen ;
4. une décision : si le CA dépasse 100 000 FCFA, afficher “Bonne journée”. Sinon afficher “Journée à renforcer”.


```python
ventes = [35000, 42000, 18000, 27000]
ca = sum(ventes)
panier_moyen = ca / len(ventes)

print("CA :", ca, "FCFA")
print("Panier moyen :", panier_moyen, "FCFA")

if ca > 100000:
    print("Bonne journée")
else:
    print("Journée à renforcer")
```
