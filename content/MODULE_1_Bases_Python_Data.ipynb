## 11. Environnement virtuel `venv`

Un environnement virtuel est un espace isolé pour un projet Python.

| Image | Explication |
|---|---|
| Ton ordinateur | La maison complète. |
| Un projet | Une chambre. |
| `venv` | Une boîte fermée avec les outils de cette chambre. |

Pourquoi ? Éviter de mélanger les bibliothèques, limiter les conflits de versions, garder un projet propre et travailler comme un vrai analyste data.

## 12. Créer et activer `venv`

### Windows

```powershell
cd C:\KORYXA_FORMATION\projets\analyse_ventes
python -m venv .venv
.venv\Scripts\activate
```

### macOS / Linux

```bash
cd ~/KORYXA_FORMATION/projets/analyse_ventes
python3 -m venv .venv
source .venv/bin/activate
```

> **Validation** : si `(.venv)` apparaît, l’environnement est actif.

## 16. Mini-projet final : analyse_ventes
Structure : main.py, notebook.ipynb, ventes.csv.
CSV : produit,montant / Formation Python,29000 / Coaching Data,45000 / Template Excel,12000 / Formation Python,29000.
Objectif : lire les ventes, calculer le CA, compter les ventes et calculer le panier moyen.


````python
import pandas as pd
ventes = pd.read_csv("ventes.csv")
ca = ventes["montant"].sum()
n = len(ventes)
panier = ca / n
print("Chiffre d affaires :", ca, "FCFA")
print("Nombre de ventes :", n)
print("Panier moyen :", panier, "FCFA")
````

## 17. Lire le résultat comme un analyste
- Chiffre d affaires : total vendu.
- Nombre de ventes : volume de transactions.
- Panier moyen : montant moyen par vente.
