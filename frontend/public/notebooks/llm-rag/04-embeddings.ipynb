# Laboratoire embeddings
Observer la similarité entre vecteurs avec un exemple local et reproductible.


```python
from math import sqrt

def cosine(a, b):
    dot = sum(x*y for x,y in zip(a,b))
    na = sqrt(sum(x*x for x in a)); nb = sqrt(sum(y*y for y in b))
    return dot/(na*nb)

question = [0.9, 0.1, 0.2]
documents = {'rag':[0.88,0.12,0.19], 'cuisine':[0.05,0.92,0.1]}
for name, vector in documents.items():
    print(name, round(cosine(question, vector), 3))
```
