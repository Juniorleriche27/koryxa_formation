# Laboratoire retrieval
Comparer top-k, seuil et bruit dans les passages récupérés.


```python
candidates = [
 ('passage A', .93, True), ('passage B', .84, True),
 ('passage C', .72, False), ('passage D', .51, False)
]
for top_k in (1,2,3,4):
    selected = candidates[:top_k]
    precision = sum(1 for _,_,relevant in selected if relevant)/len(selected)
    print('top_k=', top_k, 'precision=', precision)
```
