# Laboratoire recherche vectorielle
Simuler un top-k et inspecter les métadonnées retournées.


```python
points = [
 {'id':1,'score':0.92,'source':'guide.pdf'},
 {'id':2,'score':0.81,'source':'faq.md'},
 {'id':3,'score':0.45,'source':'notes.txt'},
]
threshold = 0.7
results = [p for p in sorted(points, key=lambda p:p['score'], reverse=True) if p['score'] >= threshold][:2]
results
```
