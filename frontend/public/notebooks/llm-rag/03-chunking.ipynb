# Laboratoire chunking
Comparer plusieurs tailles de chunks et mesurer leur lisibilité.


```python
text = 'Le RAG combine recherche et génération. ' * 40

def chunk(text, size=120, overlap=20):
    step = max(1, size-overlap)
    return [text[i:i+size] for i in range(0, len(text), step)]

for size in (80, 120, 180):
    chunks = chunk(text, size=size, overlap=20)
    print(size, len(chunks), chunks[0][:60])
```
