# Laboratoire évaluation RAG
Calculer des indicateurs simples de fidélité, pertinence et citation.


```python
tests = [
 {'faithful':1,'relevant':1,'cited':1},
 {'faithful':1,'relevant':0,'cited':1},
 {'faithful':0,'relevant':1,'cited':0},
]
for metric in ('faithful','relevant','cited'):
    score = sum(t[metric] for t in tests)/len(tests)
    print(metric, round(score, 2))
```
